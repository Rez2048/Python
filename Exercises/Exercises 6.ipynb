### 6. با استفاده از generator ها ، اعداد زی ر را تولید کرد نمایش دهید.

```python
def myrange(start,stop,step):
    while start<stop:
        yield start
        start +=step
```

### A  . اعداد بین 0 تا 1 با فاصله 0.1 و نمایش تا دو رقم اعشار:

```python
count=0
for i in myrange(0.0,1.0,0.1):
    print("%.2f" %i , end="\t")
    count+=1
    if count==5:
        print()
        count=0
```
